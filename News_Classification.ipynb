```python
import pandas as pd
d1=pd.read_csv('/content/inshort_news_data-1.csv')
d2=pd.read_csv('/content/inshort_news_data-2.csv')
d3=pd.read_csv('/content/inshort_news_data-3.csv')
d4=pd.read_csv('/content/inshort_news_data-4.csv')
d5=pd.read_csv('/content/inshort_news_data-6.csv')
```

```python
d1.head()
```

```python
def remove_others(df):
    return df[df['news_category'] != 'Others']

cleaned_d1 = remove_others(d1)
cleaned_d2 = remove_others(d2)
cleaned_d3 = remove_others(d3)
cleaned_d4 = remove_others(d4)
cleaned_d5 = remove_others(d5)

# Step 2: Merge the cleaned DataFrames
data = pd.concat([cleaned_d1, cleaned_d2, cleaned_d3, cleaned_d4, cleaned_d5], ignore_index=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sentence_transformers import SentenceTransformer
```

```python
train_df, test_df = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# Load Universal Sentence Encoder model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Embed training data
train_embeddings = model.encode(train_df['news_headline'].tolist())

# Embed testing data
test_embeddings = model.encode(test_df['news_headline'].tolist())
```

```python
# Initialize SVM classifier
svm_classifier = SVC(kernel='rbf')

# Train the SVM classifier
svm_classifier.fit(train_embeddings, train_df['label'])

# Predict labels on the test set
svm_predictions = svm_classifier.predict(test_embeddings)

# Evaluate accuracy
svm_accuracy = accuracy_score(test_df['label'], svm_predictions)
print(f"SVM Accuracy: {svm_accuracy}")
```

```python
svm_classifier = SVC(kernel='linear')

# Train the SVM classifier
svm_classifier.fit(train_embeddings, train_df['label'])

# Predict labels on the test set
svm_predictions = svm_classifier.predict(test_embeddings)

# Evaluate accuracy
svm_accuracy = accuracy_score(test_df['label'], svm_predictions)
print(f"SVM Accuracy: {svm_accuracy}")
```

```python
# Initialize KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the KNN classifier
knn_classifier.fit(train_embeddings, train_df['label'])

# Predict labels on the test set
knn_predictions = knn_classifier.predict(test_embeddings)

# Evaluate accuracy
knn_accuracy = accuracy_score(test_df['label'], knn_predictions)
print(f"KNN Accuracy: {knn_accuracy}")
```
